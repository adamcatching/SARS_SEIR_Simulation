```python
# Plot scenario automatically
def plot_infection(df):
    time = df['time']
    total_people = np.ones(len(time)) * 200
    infected = df['infected']
    recovered = df['recovered']
    infect_recover = infected + recovered
    fig = plt.figure()

    _ = plt.fill_between(time, total_people, color='navy', label='susceptible')
    _ = plt.fill_between(time, infect_recover, color='darkgreen', label='recovered')
    _ = plt.fill_between(time, infected, color='firebrick', label='infected')

    _ = plt.ylabel('number of agents')
    _ = plt.xlabel('time')
    _ = plt.title('Baseline, no Masks or Sheltering')
    _ = plt.legend()
    return fig
```

```python
def which_condition(mask=0.0, shelter=0.0, df=sim_data_df):
    per_shelt = df['percent sheltering'] == shelter
    per_mask = df['percent wearing masks'] == mask
    temp_df = df[(per_shelt) & (per_mask)]
    return temp_df
```

```python
x = plot_infection(which_condition(.2, 0.0))
```

```python
with sns.axes_style('darkgrid'):
    fig, ax = plt.subplots(10, 10, figsize=(100, 100))
    for i in range(10):
        for j in range(10):
            temp_mask = round(i * .1, 1)
            temp_shelter = round(j * .1, 1)
            temp_df = which_condition(temp_mask, temp_shelter)
            
            time = temp_df['time']
            total_people = np.ones(len(time)) * 200
            infected = temp_df['infected']
            recovered = temp_df['recovered']
            infect_recover = infected + recovered
            _ = ax[i, j].fill_between(time, total_people, color='navy', label='susceptible')
            _ = ax[i, j].fill_between(time, infect_recover, color='darkgreen', label='recovered')
            _ = ax[i, j].fill_between(time, infected, color='firebrick', label='infected')

            _ = ax[i, j].set_ylabel('number of agents')
            _ = ax[i, j].set_xlabel('time')
            _ = ax[i, j].set_title(f'{i}0% Wearing Masks, {j}0% Sheltering-in-Place')
plt.savefig('initial_simulations.png', dpi=300)
plt.show()
```

```python
# Create a map for number of people infected

# Initialize map
max_infected = np.zeros((10, 10))
for i in range(10):
        for j in range(10):
            temp_mask = round(i * .1, 1)
            temp_shelter = round(j * .1, 1)
            temp_df = which_condition(temp_mask, temp_shelter)
            last_time_point = temp_df[temp_df['time'] == 999]
            num_infected = int(last_time_point['infected'] + last_time_point['recovered'])
            max_infected[i, j] = num_infected
```

```python
# Fixing random state for reproducibility
np.random.seed(100)

# Figure ratio
fig_rat = .4
# Number of steps
N_steps = 240
# Number of agents
n_agents = int(2000 * round(fig_rat ** 2, 2))
# Set delta t
dt = .1
# Set velocity
velo = 2
# Set percent of sheltering in place
per_shelt = 0
# Set boundaries
xboundaries = [0, 100 * fig_rat]
yboundaries = [0, 100 * fig_rat]
xmax, ymax = xboundaries[1]+.5 , yboundaries[1]+.5

# Define the figure
fig = plt.figure(figsize=(10 * fig_rat, 10 * fig_rat))
ax = plt.axes(xlim=(-.5, xmax), ylim=(-.5, ymax))
plt.xticks([])
plt.yticks([])


# lists to store x and y axis points 
agents = np.zeros(n_agents, dtype=[('position', float, 2),
                                  ('velocity', float, 2),
                                  ('color', str, 20),
                                  ('state', str, 20),
                                  ('time', int, 1),
                                  ('touch', int, 1),
                                  ('num infected', int, 1)])

# Define the initial positions
agents['position'] = np.random.uniform(0, xmax, (n_agents, 2))

# Define the initial velocity angles
init_angles = np.random.uniform(0, 2 * np.pi, n_agents)
# Get initial x and y velocities from angles
velocities = np.array([[np.sin(x), np.cos(x)] for x in init_angles]) * velo
# Get percentage of agents 'sheltering'
num_shelt = int(per_shelt * n_agents)
# Make sheltering agents static
velocities[:num_shelt] = np.zeros((num_shelt, 2))
# Number of cycles for infection
asympt_infectious = 24 * 5.1

# Assign values to agents
agents['velocity'] = velocities
agents['color'] = np.array(['Navy'] * n_agents)
agents['state'] = np.array(['Susceptible'] * n_agents)
agents['time'] = np.array([0] * n_agents)
agents['num infected'] = np.zeros(n_agents)

# Make the one agent infected
agents['color'][0] = 'Firebrick'
agents['state'][0] = 'Infected'
agents['time'][0] = 1

# Initialize the plot
scatter = ax.scatter(agents['position'][:, 0], agents['position'][:, 1], 
                     color=agents['color'], s=8)
# Initialize lists storing number of infected and recovered
num_infected = [sum(agents['state']=='Infected')]
num_recovered = [0]

def update(i):
    print(i)
    # Define the new time
    t = i * dt + dt
    # Store all interactions
    interactions = np.zeros(n_agents)
    
    # Iterate over all agents 
    for j in range(n_agents):
        if agents['time'][j] != 0:
            if agents['state'][j] == 'Infected' and agents['time'][j] >= asympt_infectious:
                agents['state'][j] = 'Recovered'
                agents['color'][j] = 'Darkgreen'
                agents['time'][j] = 0
            elif agents['state'][j] == 'Infected':
                agents['time'][j] += 1
        # New displacements (prior velocity times change in time)
        deltax_temp = dt * agents['velocity'][j, 0]
        deltay_temp = dt * agents['velocity'][j, 1]
        # New positions
        x = agents['position'][j, 0] + deltax_temp
        y = agents['position'][j, 1] + deltay_temp
        # Compute new velocity
        vx_temp = deltax_temp / dt
        vy_temp = deltay_temp / dt
        
        # Add interaction
        for k in range(n_agents):
            # Go through all other agents that could be interacted with
            if interactions[k] == 0:
                if k != j:
                    # Compute components of distance
                    dx = agents['position'][j, 0] - agents['position'][k, 0]
                    dy = agents['position'][j, 1] - agents['position'][k, 1]
                    # Compute distance between agents
                    dist_temp = np.sqrt(dx ** 2 + dy ** 2)
                    # If distance of interaction is closer than both radii, collide
                    if dist_temp <= 1:
                        agents['touch'][j] += 1
                        if agents['state'][j] == 'Infected' and agents['state'][k] == 'Susceptible':
                            agents['state'][k] = 'Infected'
                            agents['color'][k] = 'Firebrick'
                            agents['time'][k] = 1
                            agents['num infected'][j] += 1
                        elif agents['state'][k] == 'Infected' and agents['state'][j] == 'Susceptible':
                            agents['state'][j] = 'Infected'
                            agents['color'][j] = 'Firebrick'
                            agents['time'][j] = 1
                            agents['num infected'][k] += 1
        if x >= xmax:
            x = 2 * xmax - deltax_temp - agents['position'][j, 0]
            vx_temp *= -1
        elif x <= 0:
            x = - deltax_temp - agents['position'][j, 0]
            vx_temp *= -1
        elif y >= ymax:
            y = 2 * ymax - deltay_temp - agents['position'][j, 1]
            vy_temp *= -1
        elif y <= 0:
            y = - deltay_temp - agents['position'][j, 1]
            vy_temp *= -1
        # Pick a new position for oldest rain drop, resetting its size,
        # color and growth factor.
        
        agents['position'][j] = np.array([x, y])
        agents['velocity'][j, 0] = vx_temp
        agents['velocity'][j, 1] = vy_temp
        interactions[j] = 1
        # Update the scatter collection, with the new colors, sizes and positions.
        scatter.set_color(agents['color'])
        scatter.set_offsets(agents['position'])
    temp = sum(agents['state']=='Infected')
    num_infected.append(temp)
    temp = sum(agents['state']=='Recovered')
    num_recovered.append(temp)
animation = FuncAnimation(fig, update, frames=N_steps, interval=24)

animation.save('interaction_test_1.gif', writer='imagemagick', dpi=100)
plt.show()
```
