```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
plt.style.use('seaborn-pastel')
```

```python
# Define the figure
fig = plt.figure()
ax = plt.axes(xlim=(0, 4), ylim=(-2, 2))
line, = ax.plot([], [], lw=3, color='red')

def init():
    line.set_data([], [])
    return line,
def animate(i):
    x = np.linspace(0, 4, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return line,

anim = FuncAnimation(fig, animate, init_func=init,
                               frames=200, interval=20, blit=True)


anim.save('sine_wave.gif', writer='imagemagick')
```

```python
# Define the figure
fig = plt.figure()
ax = plt.axes(xlim=(0, 4), ylim=(-2, 2))
line, = ax.plot([], [], lw=3, color='dodgerblue', marker='.', linestyle='-', markersize=16)

# Number of steps
N_steps = 200
# Number of agents
n_agents = 2

# lists to store x and y axis points 
xdata, ydata = [[], []] * n_agents
xdata.append(2)
ydata.append(0)

# lists to store velocity
vxdata, vydata = [[], []] * n_agents
vxdata.append(.25)
vydata.append(.5)

# Initialize the time

def init():
    line.set_data([], [])
    return line,
def animate(i):
    # Define the new time
    t = i * .1 + .1
    
    # New displacements
    deltax_temp = .1 * vxdata[i]
    deltay_temp = .1 * vydata[i]
    x = xdata[i] + deltax_temp
    y = ydata[i] + deltay_temp
    #print(t, deltax_temp, deltay_temp)
    # Compute new velocity
    vx_temp = deltax_temp / .1
    vy_temp = deltay_temp / .1
    
    if y > 2:
        y = 2 * 2 - deltay_temp - ydata[i]
        vy_temp *= -1
    elif y < -2:
        y = 2 * (-2) - deltay_temp - ydata[i]
        vy_temp *= -1
    elif x > 4:
        x = 2 * 4 - deltax_temp - xdata[i]
        vx_temp *= -1
    elif x < 0:
        x = -deltax_temp - xdata[i]
        vx_temp *= -1
    
    xdata.append(x)
    ydata.append(y)
    vxdata.append(vx_temp)
    vydata.append(vy_temp)

    
    line.set_data(xdata[i], ydata[i]) 
    return line,

anim = FuncAnimation(fig, animate, init_func=init, frames=401, interval=20, blit=True)


anim.save('test1.gif', writer='imagemagick')
```

```python
# Fixing random state for reproducibility
np.random.seed(123)

# Number of steps
N_steps = 200
# Number of agents
n_agents = 50
# Set delta t
dt = .1
# Set velocity
velo = .5
# Set boundaries
xboundaries = [0, 4]
yboundaries = [0, 4]
xmax, ymax = xboundaries[1], yboundaries[1]

# Define the figure
fig = plt.figure()
ax = plt.axes(xlim=(0, xmax), ylim=(0, ymax))

# lists to store x and y axis points 
agents = np.zeros(n_agents, dtype=[('position', float, 2),
                                  ('velocity', float, 2),
                                  ('color', str, 20)])

# Define the initial positions
agents['position'] = np.random.uniform(0, 4, (n_agents, 2))
# Define the initial angles
init_angles = np.random.uniform(0, 2 * np.pi, n_agents)
# Get initial x and y velocities from angles
velocities = np.array([[np.sin(x), np.cos(x)] for x in init_angles]) * velo
agents['velocity'] = velocities
agents['color'] = np.array(['Navy'] * n_agents)
# Initialize the plot
scat = ax.scatter(agents['position'][:, 0], agents['position'][:, 1], color=agents['color'])
all_positions = []
def update(i):
    # Define the new time
    t = i * dt + dt
    for j in range(n_agents):
        
        # New displacements (prior velocity times change in time)
        deltax_temp = dt * agents['velocity'][j, 0]
        deltay_temp = dt * agents['velocity'][j, 1]
        # New positions
        x = agents['position'][j, 0] + deltax_temp
        y = agents['position'][j, 1] + deltay_temp
        # Compute new velocity
        vx_temp = deltax_temp / dt
        vy_temp = deltay_temp / dt
        
        # 
        if x >= xmax:
            x = 2 * xmax - deltax_temp - agents['position'][j, 0]
            vx_temp *= -1
        elif x <= 0:
            x = - deltax_temp - agents['position'][j, 0]
            vx_temp *= -1
        elif y >= ymax:
            y = 2 * ymax - deltay_temp - agents['position'][j, 1]
            vy_temp *= -1
        elif y <= 0:
            y = - deltay_temp - agents['position'][j, 1]
            vy_temp *= -1
        # Pick a new position for oldest rain drop, resetting its size,
        # color and growth factor.
        
        agents['position'][j] = np.array([x, y])
        agents['velocity'][j, 0] = vx_temp
        agents['velocity'][j, 1] = vy_temp

        # Update the scatter collection, with the new colors, sizes and positions.
        scat.set_edgecolors(agents['color'])
        scat.set_offsets(agents['position'])
    temp = list(agents['position'][:])
    all_positions.append(temp)
animation = FuncAnimation(fig, update, frames=200, interval=20)

animation.save('test2.gif', writer='imagemagick')
plt.show()
```

```python
# Fixing random state for reproducibility
np.random.seed(42)

# Number of steps
N_steps = 200
# Number of agents
n_agents = 50
# Set delta t
dt = .1
# Set velocity
velo = .25
# Set boundaries
xboundaries = [0, 4]
yboundaries = [0, 4]
xmax, ymax = xboundaries[1], yboundaries[1]

# Define the figure
fig = plt.figure()
ax = plt.axes(xlim=(0, xmax), ylim=(0, ymax))

# lists to store x and y axis points 
agents = np.zeros(n_agents, dtype=[('position', float, 2),
                                  ('velocity', float, 2),
                                  ('color', str, 20),
                                  ('state', str, 20)])

# Define the initial positions
agents['position'] = np.random.uniform(0, 4, (n_agents, 2))
# Define the initial angles
init_angles = np.random.uniform(0, 2 * np.pi, n_agents)
# Get initial x and y velocities from angles
velocities = np.array([[np.sin(x), np.cos(x)] for x in init_angles]) * velo
agents['velocity'] = velocities
agents['color'] = np.array(['Navy'] * n_agents)
agents['state'] = np.array(['Susceptible'] * n_agents)

# Make the one infected
agents['color'][0] = 'Firebrick'
agents['state'][0] = 'Infected'
# Initialize the plot
scat = ax.scatter(agents['position'][:, 0], agents['position'][:, 1], color=agents['color'])
all_positions = []
def update(i):
    # Define the new time
    t = i * dt + dt
    for j in range(n_agents):
        
        # New displacements (prior velocity times change in time)
        deltax_temp = dt * agents['velocity'][j, 0]
        deltay_temp = dt * agents['velocity'][j, 1]
        # New positions
        x = agents['position'][j, 0] + deltax_temp
        y = agents['position'][j, 1] + deltay_temp
        # Compute new velocity
        vx_temp = deltax_temp / dt
        vy_temp = deltay_temp / dt
        
        # Add collision
        for k in range(n_agents):
            if k != j:
                dx = agents['position'][j, 0] - agents['position'][k, 0]
                dy = agents['position'][j, 1] - agents['position'][k, 1]
                dist_temp = np.sqrt(dx ** 2 + dy ** 2)
                # If distance of interaction is closer than both radii, collide
                if dist_temp <= .05:
                    # Get agent j velocity, velocity components, and velocity angle
                    vx_1 = agents['velocity'][j, 0]
                    vy_1 = agents['velocity'][j, 1]
                    v1 = np.sqrt(vx_1 ** 2 + vy_1 ** 2)
                    theta1 = np.arctan(vy_1 / vx_1)
                    # Get agent k velocity, velocity components, and velocity angle
                    vx_2 = agents['velocity'][k, 0]
                    vy_2 = agents['velocity'][k, 1]
                    v2 = np.sqrt(vx_2 ** 2 + vy_2 ** 2)
                    theta2 = np.arctan(vy_2 / vx_2)
                    # Get angle of contact
                    phi = np.arctan(dy / dx)
                    # Get the new velocity components of agent j
                    vx_temp = v2 * np.cos(theta2 - phi) * np.cos(phi) + v1 * np.sin(theta1 - phi) * np.cos(phi + (np.pi / 2))
                    vy_temp = v2 * np.cos(theta2 - phi) * np.sin(phi) + v1 * np.sin(theta1 - phi) * np.sin(phi + (np.pi / 2))
                    # Get the new velocity compenents of agent k
                    vx_temp_k = v1 * np.cos(theta1 - phi) * np.cos(phi) + v2 * np.sin(theta2 - phi) * np.cos(phi + (np.pi / 2))
                    vy_temp_k = v1 * np.cos(theta1 - phi) * np.sin(phi) + v2 * np.sin(theta2 - phi) * np.sin(phi + (np.pi / 2))
                    agents['velocity'][k, 0] = vx_temp_k
                    agents['velocity'][k, 1] = vy_temp_k
        if x >= xmax:
            x = 2 * xmax - deltax_temp - agents['position'][j, 0]
            vx_temp *= -1
        elif x <= 0:
            x = - deltax_temp - agents['position'][j, 0]
            vx_temp *= -1
        elif y >= ymax:
            y = 2 * ymax - deltay_temp - agents['position'][j, 1]
            vy_temp *= -1
        elif y <= 0:
            y = - deltay_temp - agents['position'][j, 1]
            vy_temp *= -1
        # Pick a new position for oldest rain drop, resetting its size,
        # color and growth factor.
        
        agents['position'][j] = np.array([x, y])
        agents['velocity'][j, 0] = vx_temp
        agents['velocity'][j, 1] = vy_temp

        # Update the scatter collection, with the new colors, sizes and positions.
        scat.set_edgecolors(agents['color'])
        scat.set_offsets(agents['position'])
    temp = list(agents['position'][:])
    all_positions.append(temp)
animation = FuncAnimation(fig, update, frames=200, interval=20)

animation.save('test3.gif', writer='imagemagick')
plt.show()
```

```python
# import seaborn as sns
sns.set(style="ticks")

# Fixing random state for reproducibility
np.random.seed(100)

# Number of steps
N_steps = 200
# Number of agents
n_agents = 100
# Set delta t
dt = .1
# Set velocity
velo = .2
# Set boundaries
xboundaries = [0, 4]
yboundaries = [0, 4]
xmax, ymax = xboundaries[1], yboundaries[1]

# Define the figure
fig = plt.figure()
ax = plt.axes(xlim=(0, xmax), ylim=(0, ymax))

# lists to store x and y axis points 
agents = np.zeros(n_agents, dtype=[('position', float, 2),
                                  ('velocity', float, 2),
                                  ('color', str, 20),
                                  ('state', str, 20),
                                  ('time', int, 1)])

# Define the initial positions
agents['position'] = np.random.uniform(0, 4, (n_agents, 2))
# Define the initial angles
init_angles = np.random.uniform(0, 2 * np.pi, n_agents)
# Get initial x and y velocities from angles
velocities = np.array([[np.sin(x), np.cos(x)] for x in init_angles]) * velo
agents['velocity'] = velocities
agents['color'] = np.array(['Navy'] * n_agents)
agents['state'] = np.array(['Susceptible'] * n_agents)
agents['time'] = np.array([0] * n_agents)

# Make the one infected
agents['color'][0] = 'Firebrick'
agents['state'][0] = 'Infected'
agents['time'][0] = 1
# Initialize the plot
scat = ax.scatter(agents['position'][:, 0], agents['position'][:, 1], color=agents['color'])
num_infected = [sum(agents['state']=='Infected')]
num_recovered = [0]
def update(i):
    # Define the new time
    t = i * dt + dt
    for j in range(n_agents):
        if agents['time'][j] != 0:
            if agents['state'][j] == 'Infected' and agents['time'][j] >= 100:
                agents['state'][j] = 'Recovered'
                agents['color'][j] = 'Darkgreen'
                agents['time'][j] = 0
            elif agents['state'][j] == 'Infected':
                agents['time'][j] += 1
        # New displacements (prior velocity times change in time)
        deltax_temp = dt * agents['velocity'][j, 0]
        deltay_temp = dt * agents['velocity'][j, 1]
        # New positions
        x = agents['position'][j, 0] + deltax_temp
        y = agents['position'][j, 1] + deltay_temp
        # Compute new velocity
        vx_temp = deltax_temp / dt
        vy_temp = deltay_temp / dt
        
        # Add collision
        for k in range(n_agents):
            # Go through all other agents that could be interacted with
            if k != j:
                # Compute components of distance
                dx = agents['position'][j, 0] - agents['position'][k, 0]
                dy = agents['position'][j, 1] - agents['position'][k, 1]
                # Compute distance between agents
                dist_temp = np.sqrt(dx ** 2 + dy ** 2)
                # If distance of interaction is closer than both radii, collide
                if dist_temp <= .05:
                    if agents['state'][j] == 'Infected':
                        agents['state'][k] = 'Infected'
                        agents['time'][k] += 1
                        agents['color'][k] = 'Firebrick'
        if x >= xmax:
            x = 2 * xmax - deltax_temp - agents['position'][j, 0]
            vx_temp *= -1
        elif x <= 0:
            x = - deltax_temp - agents['position'][j, 0]
            vx_temp *= -1
        elif y >= ymax:
            y = 2 * ymax - deltay_temp - agents['position'][j, 1]
            vy_temp *= -1
        elif y <= 0:
            y = - deltay_temp - agents['position'][j, 1]
            vy_temp *= -1
        # Pick a new position for oldest rain drop, resetting its size,
        # color and growth factor.
        
        agents['position'][j] = np.array([x, y])
        agents['velocity'][j, 0] = vx_temp
        agents['velocity'][j, 1] = vy_temp

        # Update the scatter collection, with the new colors, sizes and positions.
        scat.set_color(agents['color'])
        scat.set_offsets(agents['position'])
    temp = sum(agents['state']=='Infected')
    num_infected.append(temp)
    temp = sum(agents['state']=='Recovered')
    num_recovered.append(temp)
animation = FuncAnimation(fig, update, frames=1000, interval=20)

animation.save('test6.gif', writer='imagemagick')
plt.show()
```
